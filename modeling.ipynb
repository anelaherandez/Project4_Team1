```python
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import warnings
 
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import KNNImputer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
```

```python
Zillow_analysis = Zillow.drop(columns = ['Street', 'State','Latitude', 'Longitude', 'LotArea', 'Unnamed: 0'])
```

```python
Zillow_analysis.head()
```

```python
X = Zillow_analysis.iloc[:,1:7].values  #features
y = Zillow_analysis.iloc[:,7].values
```

```python

#help from: https://www.geeksforgeeks.org/random-forest-regression-in-python/
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
 
label_encoder = LabelEncoder()
x_categorical = Zillow_analysis.select_dtypes(include=['object']).apply(label_encoder.fit_transform)
x_numerical = Zillow_analysis.select_dtypes(exclude=['object']).values
x = pd.concat([pd.DataFrame(x_numerical), x_categorical], axis=1).values
 
# Fitting Random Forest Regression to the dataset
regressor = RandomForestRegressor(n_estimators=10, random_state=0, oob_score=True)
 
# Fit the regressor with x and y data
regressor.fit(x, y)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
 
# Access the OOB Score
oob_score = regressor.oob_score_
print(f'Out-of-Bag Score: {oob_score}')
 
# Making predictions on the same data or new data
predictions = regressor.predict(x)
 
# Evaluating the model
mse = mean_squared_error(y, predictions)
print(f'Mean Squared Error: {mse}')
 
r2 = r2_score(y, predictions)
print(f'R-squared: {r2}') #Random forrest model has accuracy R-squared of over 0.998
```
