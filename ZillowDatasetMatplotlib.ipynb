```python
bed_size= ("2 bedroom", "3 Bedroom", "4 Bedroom")
cities = ("Tucson", "Prescott", "Scottsdale", "Phoenix")
highest_valued = {
    'Tucson': (1610, 1850, 775),
    'Prescott': (675, 975, 2499),
    'Scottsdale': (710, 1250, 2999),
    'Phoenix' : (1289, 3000, 1599)
}

x = np.arange(len(bed_size))  # the label locations
width = 0.15  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in highest_valued.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Length (mm)')
ax.set_title('Highest Valued 2,3 and 4 Bedrooms')
ax.set_xticks(x + width, bed_size)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(500, 3500)

plt.show()
```
